**Answer:** :

2. For the quantitative data, the probability distribution of a random variable is defined by a probability density function, f(x). Exp 1: 10 independent throws

```r
exp1 <- as.integer(runif(10, 1, 7))
plot(table(exp1))
```

Exp 2: 1,000 independent throws
