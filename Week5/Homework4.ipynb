1. Using the mean squared error (MSE) as your objective function allows for a closed form solution to finding the maximum likelihood estimate (MLE) of your model parameters in linear regression. Let’s consider the simple, single predictor variable model $Y= \beta_0 + \beta_1 X $. 

a) Use algebra to show how you can expand out the MSE to get from i to ii below. Do your proof using Latex equation formatting.


 _i)_ $MSE(\beta_0, \beta_1)=E[ (Y-(\beta_0 + \beta_1 X))^2]$ 
 
 _ii)_ $MSE(\beta_0, \beta_1)=E[Y^2] -2 \beta_0E[Y]-2 \beta_1 Cov[X,Y]-2 \beta_1 E[X]E[Y]+ \beta_0^2 +2 \beta_0 \beta_1 E[X]+\beta_1^2 Var[X]+ \beta_1^2 (E[X])^2$

**Answer:**



<font color='blue'>
\begin{aligned}
 MSE(\beta_0, \beta_1) &= E[ (Y-(\beta_0 + \beta_1 X))^2] \\
  &= E[Y^2 + (\beta_0 + \beta_1 X) ^2 - 2Y(\beta_0 + \beta_1 X)] \\
  &= E[Y^2 + \beta_0^2 + \beta_1^2X^2 + 2X\beta_0\beta_1 - 2Y\beta_0 - 2XY\beta_1] \\
  &= E[Y^2] + \beta_0^2 + \beta_1^2E[X^2] + 2\beta_0\beta_1E[X] - 2\beta_0E[Y] - 2\beta_1E[XY] \\
  &= E[Y^2] + \beta_0^2 + \beta_1^2(Var[X] + E[X]^2) + 2\beta_0\beta_1E[X] - 2\beta_0E[Y] - 2\beta_1(Cov[X,Y] + E[X]E[Y]) \\
  &= E[Y^2] + \beta_0^2 + \beta_1^2Var[X] + \beta_1^2E[X]^2 + 2\beta_0\beta_1E[X] - 2\beta_0E[Y] - 2\beta_1Cov[X,Y] - 2\beta_1E[X]E[Y] \\
  & = E[Y^2] - 2\beta_0E[Y] - 2\beta_1Cov[X,Y] - 2\beta_1E[X]E[Y] + \beta_0^2 + 2\beta_0\beta_1E[X] + \beta_1^2Var[X] + \beta_1^2(E[X])^2
\end{aligned}
<font>

b) Prove that the MLE of $\beta_0$ is $E[Y]- \beta_1 E[X]$ by taking the derivative of _ii_ above, with respect to $\beta_0$, setting the derivative to zero, and solving for $\beta_0$.

**Answer:**
<font color='blue'>
$$\frac{\partial MSE(\beta_0, \beta_1)}{\partial \beta_0} = \frac{\partial E[Y^2] - 2\beta_0E[Y] - 2\beta_1Cov[X,Y] - 2\beta_1E[X]E[Y] + \beta_0^2 + 2\beta_0\beta_1E[X] + \beta_1^2Var[X] + \beta_1^2(E[X])^2}{\partial \beta_0}  $$


Setting the derivative to 0 : 

\begin{aligned}
 0 &= \frac{\partial E[Y^2] - 2\beta_0E[Y] - 2\beta_1Cov[X,Y] - 2\beta_1E[X]E[Y] + \beta_0^2 + 2\beta_0\beta_1E[X] + \beta_1^2Var[X] + \beta_1^2(E[X])^2}{\partial \beta_0} \\
 0 &= -2E[Y] + 2\beta_0 + 2\beta_1E[X] \\
 0 &= E[Y] - \beta_0 - \beta_1E[X] \\
 \beta_0 &= E[Y] - \beta_1E[X]
\end{aligned}
<font>
      

c) Prove that the MLE for $\beta_1$ is $Cov[X,Y]/Var[X]$ by taking the derivative of _ii_ above, with respect to $\beta_1$, setting the derivative to zero, and solving for $\beta_1$. 

**Answer:**
<font color = "blue">
$$\frac{\partial MSE(\beta_0, \beta_1)}{\partial \beta_1} = \frac{\partial E[Y^2] - 2\beta_0E[Y] - 2\beta_1Cov[X,Y] - 2\beta_1E[X]E[Y] + \beta_0^2 + 2\beta_0\beta_1E[X] + \beta_1^2Var[X] + \beta_1^2(E[X])^2}{\partial \beta_1}  $$

Setting the derivatie to 0: 

\begin{aligned}
 0 &= \frac{\partial E[Y^2] - 2\beta_0E[Y] - 2\beta_1Cov[X,Y] - 2\beta_1E[X]E[Y] + \beta_0^2 + 2\beta_0\beta_1E[X] + \beta_1^2Var[X] + \beta_1^2(E[X])^2}{\partial \beta_1} \\
 0 &= -2Cov[X,Y] - 2E[X]E[Y] + 2\beta_0E[X] + 2\beta_1Var[X] + 2\beta_1(E[X])^2 \\
 0 &= Cov[X,Y] + E[X]E[Y] - \beta_0E[X] - \beta_1Var[X] - \beta_1(E[X])^2 \\
 \beta_1Var[X] + \beta_1(E[X])^2 &= Cov[X,Y] + E[X]E[Y] - \beta_0E[X] \\
 \beta_1Var[X] + \beta_1(E[X])^2 &= Cov[X,Y] + E[X]E[Y] - (E[Y] - \beta_1E[X])E[X] \\
 \beta_1Var[X] + \beta_1(E[X])^2 &= Cov[X,Y] + E[X]E[Y] - E[Y]E[X] + \beta_1(E[X])^2 \\
 \beta_1Var[X] &= Cov[X,Y] 
\end{aligned}

Therefore we have: 

$$ \beta_1 = \frac{Cov[X,Y]}{Var[X]} $$

3. In the case where X and Y are standard normal variables (i.e. $N(0,1)$), show that the ordinary least squares solution converges to the same solution as the Pearson correlation coefficient $r=Cov[X,Y]/(\sigma_X * \sigma_Y)$. What exactly do they both converge to?


**Answer:**

<font color = "blue">
When X and Y are standard normal variables, we have: 

$\sigma_X = 1, \sigma_Y = 1, E[X] = 0, E[Y] = 0$

Substitute the above value into the solutions: 

For the intercept:

\begin{aligned}
    \beta_0 &= E[Y] - \beta_1E[X] \\
    \beta_0 &= 0 - \beta_1*0 \\
    \beta_0 &= 0
\end{aligned}

For the slope:

\begin{aligned}
    \beta_1 &= \frac{Cov[X,Y]}{Var[X]} \\
    \beta_1 &= \frac{Cov[X,Y]}{sigma_X} \\
    \beta_1 &= Cov[X,Y]\\
    \beta_1 &= Cov[\frac{X-E[X]}{\sigma_X}, \frac{Y-E[Y]}{\sigma_Y}]\\
    \beta_1 &= Cov[\frac{X}{\sigma_X}, \frac{Y}{\sigma_Y}]\\
    \beta_1 &= \frac{Cov[X,Y]}{\sigma_X * \sigma_Y} = r
\end{aligned}

They both converge to the slope of a simple linear regression line.

4. Three methods for evaluating a regression model’s performance are $r^2$, $RSE$, and the $F$ Test. Show the functional form of each method and describe conceptually what it is telling you about your model.



**Answer:**
<font color = "blue">

$r^2 = \frac{TSS-RSS}{TSS}$

$RSE = \sqrt{ \frac{RSS}{n-2} } $

$F = \frac{ \frac{(TSS-RSS)}{p}} { \frac{RSS}{n - p - 1}} $

in which: 
$TSS = \sum_{i=1}^{n} (y_i - \bar{y})^2$, $RSS = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

and $n$ refers to the sample size and $p$ to the number of variables

- $RSE$
    - measures on average how much the response will deviate from the true regression line. The smaller the value, the better the fit.  
    
- $r^2$
    - measures what proportion of the variance in Y is explained by X. The closer to 1, the larger proportion of the variance in response is explained by the model. 

- $F$
    - the ratio indicates whether there is a relationship between the response and the predictor. A closer to 1 value indicates the relationship is weak, and that we can not reject the null hypothesis that all of the coefficients in our model equal to zero. Conversely, if the ratio is far greater than 1, then we are more confident to reject the null hypothesis.
